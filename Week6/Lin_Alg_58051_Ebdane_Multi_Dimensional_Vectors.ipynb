```python
import numpy as np
import matplotlib.pyplot as plt
```

$$J = 9x + 17y $$

```python
P = np.array ([9, 17])

plt.quiver(0,0, P[0], P[1], angles= 'xy', scale_units= 'xy',
              scale=1, color= 'Orange', label='P')

plt.xlim(0,15)
plt.ylim(0,15)
plt.grid()
plt.legend()
plt.show()
```

$$E = \left\{
    \begin{array}\\
        5x + 2y + 8z \\ 
        4x + 9y + 3z \\
        2x +4y + 9z
    \end{array}
\right. $$

```python
E = np.array([
              [5,2,8],
              [4,9,3],
              [2,4,9]
])
```

```python
figure = plt.figure()
axis = figure.gca(projection='3d')

plt.quiver(0,0,0, E[0,0], E[0,1], E[0,2], color= 'Green')
plt.quiver(0,0,0, E[1,0], E[1,1], E[1,2], color= 'Blue')
plt.quiver(0,0,0, E[2,0], E[2,1], E[2,2], color= 'Red')

axis.set_xlim(0,15)
axis.set_ylim(0,15)
axis.set_zlim(0,15)

origin = (0,0,0)
axis.quiver(origin, origin, origin, E[:,0], E[:,1], E[:,2], 
           arrow_length_ratio=0.05, colors=['Green','Blue','Red'])
plt.show()
```
