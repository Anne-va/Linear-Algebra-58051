```
import numpy as np
import matplotlib.pyplot as plt

profit = np.array ([1500, 2000, 1700, 3000]) 
reach = np.array ([12000, 17000, 5000, 4500])
revenue = np.array ([19000, 7500, 8000, 9000])

def month_profit_trace(profit, reach, make_figs=True): 
    if (profit.shape == (4,)) and (revenue.shape == (4,)): 
        e_week1 = np.array ((reach[0], profit[0]))
        e_week2 = np.array ((reach[1], profit[1]))
        e_week3 = np.array ((reach[2], profit[2]))
        e_week4 = np.array ((reach[3], profit[3]))
        
        e_month = np.square((np.square(e_week1, e_week2), e_week3, e_week4))
        e_month_magnitude = np.sqrt(np.sum(e_month))
        alpha = 6.31134
        theta = np.arctan(np.add(np.sin(e_month_magnitude)/np.cos(e_month_magnitude), alpha))
        theta = np.degrees(theta)  
        

        plt.figure(figsize=(16,5))
        plt.title('Bebang\'s Month Post Efficiency')
        plt.xlim(0,1.01*np.sum(reach))
        plt.ylim(-np.sum(np.abs(profit)),np.sum(np.abs(profit)))
        plt.xlabel('FB Post Reach Increment')
        plt.ylabel('Profit')
        plt.grid()
        n = 2
        
        weekly_efficiency = np.add(e_week1, e_week2)
        weekly_efficiency4 = np.add(np.add(e_week1, e_week2), e_week3, e_week4)


        plt.quiver(0, 0, e_week1[0], e_week1[1],
                   angles='xy', scale_units='xy', scale=1, color='green',
                   label='Week 1: {:.2f}'.format(np.linalg.norm(e_week1)))
        plt.quiver(e_week1[0], e_week1[1], e_week2[0], e_week2[1],
                   angles='xy', scale_units='xy', scale=1, color='pink',
                   label='Week 2: {:.2f}'.format(np.linalg.norm(e_week2)))
        plt.quiver(0, 0, weekly_efficiency[0],weekly_efficiency[1],
                   angles='xy', scale_units='xy',scale=1, color= 'violet',
                   label='Week 3: {:.2f}'.format(np.linalg.norm(e_week3)))
        plt.quiver(weekly_efficiency[0],weekly_efficiency[1], e_week3[0], e_week3[1],
                   angles='xy', scale_units='xy',scale=1, color= 'red',
                   label='Week 3: {:.2f}'.format(np.linalg.norm(e_week3)))
        plt.quiver(weekly_efficiency4[0],weekly_efficiency4[1], e_week4[0], e_week4[1],
                   angles='xy', scale_units='xy',scale=1, color= 'yellow',
                   label='Week 4: {:.2f}'.format(np.linalg.norm(e_week4)))
        plt.quiver(0, 0, weekly_efficiency4[0],weekly_efficiency4[1],
                   angles='xy', scale_units='xy',scale=1, color= 'orange',
                   label='R2: {:.2f}'.format(np.linalg.norm(weekly_efficiency4)))

        figure = plt.figure()
        axis = figure.gca(projection='3d')


        axis.quiver(0, 0, 0, e_month[0], e_month[1], e_month[2], 
                 
                 label = 'Efficiency: {:.2f} @ {:.2f}'.format(e_month_magnitude, theta))
        
        plt.legend(loc='upper left')

        if make_figs:
          plt.savefig(f'LinAlg-Lab2-Bebang Post Eff-{int(e_month_magnitude)}@{int(theta)}.png', dpi=300)
                    
        plt.show()

    else:  
        print('Dimension error')

month_profit_trace(profit, reach, make_figs=True)
```
